```python
def df_to_X_y(df, window_size = 5):
  df_as_np = df.to_numpy()
  X = []
  y = []
  for i in range(len(df_as_np)-window_size):
    row = df_as_np[i:i+window_size]
    X.append(row)
    label = df_as_np[i+window_size]
    y.append(label)
  return np.array(X), np.array(y)
```

```python
from sklearn.preprocessing import StandardScaler

features = df[['T (degC)', 'rh (%)', 'p (mbar)', 'wv (m/s)', 'Tdew (degC)']]

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

scaled_features = pd.DataFrame(scaled_features, columns=features.columns)
```

```python
WINDOW_SIZE = 72
X, y = df_to_X_y(scaled_features, WINDOW_SIZE)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, InputLayer
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.optimizers import Adam
```

```python
X_train, y_train = X[:60000], y[:60000]
X_val, y_val = X[60000:65000], y[60000:65000]
X_test, y_test = X[65000:], y[65000:]
```

```python
X_train.shape, y_train.shape
```

```python
X_val.shape, y_val.shape
```

```python
X_test.shape, y_test.shape
```

```python
model = Sequential()

model.add(InputLayer((72, 5)))
model.add(LSTM(64, dropout=0.2, return_sequences=True))
model.add(LSTM(32))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='linear'))

model.summary()
```

```python
from tensorflow.keras.metrics import RootMeanSquaredError
```

```python
cp = ModelCheckpoint("model/model.keras", save_best_only=True)
```

```python
model.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.001), metrics = [RootMeanSquaredError()])
```

```python
model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=20, batch_size=64, shuffle=False, callbacks=[cp])
```

```python
# Load best model
from tensorflow.keras.models import load_model
model = load_model('model/model.keras')

# Evaluate on test set
test_loss, test_rmse = model.evaluate(X_test, y_test)
print(f"Test RMSE (scaled): {test_rmse:.4f}")

# Inverse transform to get °C
predictions = model.predict(X_test)

# Reconstruct full feature arrays for inverse transform
pred_full = np.zeros((len(predictions), 5))
pred_full[:, 0] = predictions[:, 0]
actual_full = np.zeros((len(y_test), 5))
actual_full[:, 0] = y_test[:, 0]

pred_celsius = scaler.inverse_transform(pred_full)[:, 0]
actual_celsius = scaler.inverse_transform(actual_full)[:, 0]

mae = np.mean(np.abs(pred_celsius - actual_celsius))
print(f"Test MAE: {mae:.2f}°C")

# Plot
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 5))
plt.plot(actual_celsius[:200], label='Actual Temperature (°C)')
plt.plot(pred_celsius[:200], label='Predicted Temperature (°C)')
plt.legend()
plt.title('LSTM Temperature Forecasting — Predicted vs Actual')
plt.xlabel('Time steps')
plt.ylabel('Temperature (°C)')
plt.tight_layout()
plt.savefig('prediction_plot.png', dpi=150)
plt.show()
```
